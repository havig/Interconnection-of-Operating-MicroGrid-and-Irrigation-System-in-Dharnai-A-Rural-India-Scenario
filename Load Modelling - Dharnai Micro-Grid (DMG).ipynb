Load Modelling for the domestic load is done based on the following parameters:
1. Number of appliances in the small, medium and large houses and in hospital and school as well.
2. Usage differes on weekends and weekdays which is also considered. 
3. The population increase per year of the village is considered while modelling as well.

The code below shows how the load is modelled for the appliance bulb for the first year. The load is simulated for other appliances in a similar way for a period of 20 years.
